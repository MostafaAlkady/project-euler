```python
def is_palindrome(x):
    n = 0
    x = str(x)
    for i in range(len(x)//2):
        if x[i] == x[-1-i]:
            n += 1
    if n == len(x)//2:
        return True
    else:
        return False
```

```python
number_of_lychrel_numbers = 0
for i in range(10,10001):
    number_of_iterations = 0
    if i % 100 == 0:
        print("Now calculating:", i)
    reversed_and_added = i + int(str(i)[::-1])
    while is_palindrome(reversed_and_added) == False:
        # print(i,number_of_iterations)
        reversed_and_added_2 = reversed_and_added + int(str(reversed_and_added)[::-1])
        reversed_and_added = reversed_and_added_2
        number_of_iterations += 1
        if number_of_iterations == 50:
            number_of_lychrel_numbers += 1
            break
print(number_of_lychrel_numbers)
```
