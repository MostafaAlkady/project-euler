```python
file_path = '0042_words.txt'  # Replace with the actual path to your text file
file = open(file_path, 'r')

read = file.readlines()
read_string = read[0]
word_list = read_string.split('","')
word_list[0] = 'A'
word_list[-1] = 'YOUTH'

print(word_list)
```

```python
triangle_numbers = []
for i in range(1,200):
  triangle_numbers.append(i*(i+1)/2)
count = 0
for word in word_list:
  word_value = 0
  for char in word:
    word_value += ord(str(char))-64
  print("The word is", str(word), "value is:", word_value)
  if word_value in triangle_numbers:
    count += 1

print(count)
```
