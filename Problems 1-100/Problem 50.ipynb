```python
import numpy as np
import math
```

```python
def prime_list(m): #generate a list of m prime numbers 
    l = [2]
    n = 3
    while len(l) < m:
        for j in l:
            if n%j == 0:
                break
        else:
            l.append(n)
        n += 2
    return l
```

```python
def is_prime(n): #check if n is a prime number
    if n <= 1:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True
```

```python
x = prime_list(80000)
X = np.array(x)
```

```python
for chain_length in range(1,1000,1):
    for i in range(chain_length):
        # sum(x[i:i+chain_length])
        if is_prime(np.sum(X[i:i+chain_length])) and np.sum(X[i:i+chain_length])<1000000:
            print("sum of",chain_length,"numbers:",np.sum(x[i:i+chain_length]))
    chain_length += 2
```
