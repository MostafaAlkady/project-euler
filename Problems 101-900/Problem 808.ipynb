```python
import numpy as np
prime_numbers = []  # Declare the global variable

def sieve_of_eratosthenes(limit):
    global prime_numbers  # Declare that we are using the global variable
    prime = [True] * (limit + 1)
    p = 2

    while p * p <= limit:
        if prime[p]:
            for i in range(p * p, limit + 1, p):
                prime[i] = False
        p += 1

    prime_numbers = [p for p in range(2, limit + 1) if prime[p]]
    return prime_numbers

# Example usage
limit = 99999999
primes = sieve_of_eratosthenes(limit)
primes = np.array(primes)
# print(sieve_of_eratosthenes(limit))

def is_palindrome(n):
    return str(n) == str(n)[::-1]

def reverse_string(s):
    return s[::-1]

def is_prime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n ** 0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

def is_integer(input_str):
    try:
        # Try to convert the input to a float
        number = float(input_str)
        
        # Check if the number is an integer
        return number.is_integer()
    except ValueError:
        # If conversion to float fails, it is not a number
        return False
```

```python
primes_squared = primes ** 2
RPS = []
counter = 0
for i in primes_squared:
    reversed_i = int(reverse_string(str(i)))
    if not is_palindrome(i) and is_integer(np.sqrt(reversed_i)) and is_prime(np.sqrt(int(reverse_string(str(i))))):
        counter += 1
        print(counter, ":", i)
        RPS.append(i)
        if counter == 50:
            break

print("sum is:", np.sum(RPS))
print(len(RPS))
```
